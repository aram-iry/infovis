```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Direct-catches-summary-data-since-1985-V7_2-2025.csv', sep=';')

# Group by year and sum the total catches
yearly_totals = df.groupby('Year')['Total'].sum().reset_index()

# Create the plot
plt.figure(figsize=(12, 6))
plt.bar(yearly_totals['Year'], yearly_totals['Total'], color='skyblue')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Whale Catches')
plt.title('Total Whale Catches by Year (1985-2024)')
plt.xticks(rotation=45)
```
